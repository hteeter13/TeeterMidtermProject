```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data2016 = pd.read_excel('2016 November General Election.xlsx')
data2016.head()
```

```python
data2016.tail()
```

```python
data2016.shape
```

```python
data2016 = data2016.drop(['Status', 'State Results Website', 'Overseas Eligible', 'Total Ballots Counted (Estimate)', 'Voting-Eligible Population (VEP)', 'State Abv', 'VAP Highest Office', 'Highest Office', '% Non-citizen', 'Prison', 'Probation', 'Parole', 'Total Ineligible Felon', 'Voting-Age Population (VAP)'], axis=1)
data2016
```

```python
data2016.rename(columns={'Unnamed: 0': 'State'}, inplace=True)
data2016
```

```python
data_avr = pd.read_excel('States with AVR 2016.xlsx')
data_avr
```

```python
data_avr2016 = pd.concat([data2016, data_avr], axis=1)
data_avr2016
```

```python
data_avr2016 = data_avr2016.loc[:, ~data_avr2016.columns.duplicated()]
data_avr2016
```

```python
pd.options.display.float_format = '{:.2%}'.format
```

```python
data_avr2016
```

```python
data_avr2016.to_csv('data_avr2016.csv', index=False)
```
