```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data2020 = pd.read_excel('2020 November General Election.xlsx')
data2020.head(10)
```

```python
data2020.shape
```

```python
data2020 = data2020.drop(['Source', 'Unnamed: 2', 'Total Ballots Counted (Estimate)', 'Vote for Highest Office (President)', 'Voting-Eligible Population (VEP)', 'Voting-Age Population (VAP)', '% Non-citizen', 'Prison', 'Probation', 'Parole', 'Total Ineligible Felon', 'Overseas Eligible', 'State Abv'], axis=1)
data2020
```

```python
data2020.rename(columns={'Unnamed: 0': 'State', 'VEP Turnout Rate (Total Ballots Counted)':'VEP Total Ballots Counted', 'VEP Turnout Rate (Highest Office)':'VEP Highest Office'}, inplace=True)
data2020
```

```python
data_avr2 = pd.read_excel('States with AVR 2020.xlsx')
data_avr2
```

```python
data_avr2020 = pd.concat([data2020, data_avr2], axis=1)
data_avr2020
```

```python
data_avr2020 = data_avr2020.loc[:, ~data_avr2020.columns.duplicated()]
data_avr2020
```

```python
data_avr2020.to_csv('data_avr2020.csv', index=False)
```
